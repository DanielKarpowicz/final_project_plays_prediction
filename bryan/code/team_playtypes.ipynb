```python
import pandas as pd
```

```python
file = '../raw_data/NFL.csv'
```

```python
df = pd.read_csv(file, low_memory=False)
```

```python
df.head()
```

```python
pd.set_option('max_columns', None)
```

```python
df
```

```python
simple_df = df[(df.down.isin([1.0, 2.0, 3.0, 4.0])) & ((df.play_type=='run') | (df.play_type == 'pass'))]
simple_df.head()
```

```python
simple_df = simple_df[['game_id','posteam', 'posteam_type', 'game_seconds_remaining', 'yardline_100', 'down', 'ydstogo', 'play_type', 'pass_length', 'pass_location', 'run_location']]
simple_df
```

```python
import numpy as np
def get_full_play_type(play):
    play_type, pass_location, run_location = play
    if(play_type == 'run'):
        return play_type+'_'+ run_location
    else:
        return play_type+'_'+ pass_location
simple_df = simple_df.replace(np.nan, 'unknown', regex=True)    
simple_df['full_play_type'] = simple_df[['play_type','pass_location', 'run_location']].apply(get_full_play_type, axis=1)
simple_df = simple_df[(simple_df.full_play_type.isin(['pass_left', 'pass_middle','pass_right','run_left', 'run_middle', 'run_right']))]
```

```python
team_play_types = simple_df.groupby(['posteam', 'full_play_type']).count()[['play_type']]
```

```python
team_play_types.head(12)
```

```python
team_play_types = team_play_types.reset_index()[['posteam', 'full_play_type', 'play_type']]
team_play_types = team_play_types.rename(columns={"posteam":"Team", "play_type":"play_count"})
team_play_types
```

```python
team_play_types.to_csv("../processed_data/team_play_types.csv", index=False, header=True)
```

```python
nfl_play_types = simple_df.groupby(['full_play_type']).count()[['play_type']]
nfl_play_types
```

```python
nfl_play_types = nfl_play_types.reset_index()[['full_play_type', 'play_type']]
nfl_play_types = nfl_play_types.rename(columns={"play_type":"play_count"})
nfl_play_types
```

```python
nfl_play_types.to_csv("../processed_data/nfl_play_types.csv", index=False, header=True)
```
