```python
import pandas as pd
import matplotlib
from sklearn import preprocessing
%matplotlib inline
```

```python
df = pd.read_csv(r'C:\Users\dakar\Desktop\NFL\raw.csv', low_memory=False)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
#select important columns for input
X=df[['yardline_100', 'down', 'ydstogo', 'down_1.0', 'down_2.0', 'down_3.0', 'down_4.0','ydstogo_0','ydstogo_1','ydstogo_2','ydstogo_3','ydstogo_4']]
#select result column for output
Y=df['full_play_type_code']
#split data for train and test
train_x, test_x, train_y, test_y = train_test_split(X, Y, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeClassifier
the_clf=DecisionTreeClassifier(class_weight=None, 
max_depth=None, max_features=None, max_leaf_nodes=None, 
min_samples_leaf=1, min_samples_split=2, min_weight_fraction_leaf=0.0)
the_clf.fit(train_x, train_y)
```

```python
from sklearn.metrics import accuracy_score, balanced_accuracy_score, adjusted_mutual_info_score, explained_variance_score
pred = the_clf.predict(test_x)
acc =accuracy_score(test_y, pred)
print(acc)
```

```python
bacc = balanced_accuracy_score(test_y, pred)
print(bacc)
```

```python
macc = adjusted_mutual_info_score(test_y, pred)
print(macc)
```

```python
vacc = explained_variance_score(test_y, pred)
print(vacc)
```
